```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

## K-Means

### Importando o dataset

```python
data_quake = pd.read_csv('quake.csv', header=None).values
```

### Visualizando os dados

```python
fig = go.Figure()
fig.add_trace(go.Scattergeo(
    locationmode='country names',
    lon = data_quake[:, 1],
    lat = data_quake[:, 0],
    marker = dict(
        size=3,
        color = '#444444'
    )
))
fig.update_layout(
    title = dict(
        text = 'Localizao dos terremotos registrados',
        font = dict(size=30, color = '#444444'),
        x = 0.5,
        y = 0.9
    ),
    autosize=False,
    width=1200,
    height=800
)
```
